```python
import torch
a = torch.randn([2,3])
import torch.nn as nn

#基本上所有torch操作可以直接作用于Tensor实例

#有七种CPU Tensor和八种GPU tensor

#可以使用索引和切片来改变Tensor的值
x = torch.FloatTensor([[-1, 2, 3], [4, 5, 6]])
x[0][1] = 8

#Tensor可以记录梯度
x = torch.tensor([[1., -1.], [1., 1.]], requires_grad=True)
out = x.pow(2).sum()
out.backward()
print(x.grad)

#会改变tensor的函数操作会用一个下划线后缀来标示。比如zero_
#不带下划线的是非就地版本
#这些方法与torch库中的数学操作一致
#计算图中的叶子节点无法进行就地 in place运算
x = x.detach()
x.div_(0.5)
x.add(1) #不是in place
x.abs_()
x.zero_()
x.add_(1)
x.mm(x) #二维矩阵乘法  x.mv是与向量相乘
x.matmul(x) #更广义的矩阵乘法，支持多维张量
#x.bmm(x) #必须是3D的
x.ceil_() #向上取整
t = x.chunk(chunks=2,dim=0) #按照dim维度均分为chunks
x.clamp_(-1,1) #压缩到-1，1
x.mul(3) #按位乘法
x.prod() #返回每个元素的乘积
x.size()
x.shape #size的别名
x.split(3, 1) #在1维度按照3大小划分
x.to_sparse() #返回为稀疏存储方式的副本
x.transpose(0,1) #交换两个维度


a = torch.tensor([3+1j,4+2j,5+3j],dtype=torch.cfloat)
a.imag #返回虚部
a.diag()  #返回对角元素
a.diff() #返回相邻元素的差
a.dim() #返回维度
a.cpu() #返回一个cpu的副本
a.norm(p=1) #计算范数
a.normal_() #进行填充
a.numpy() #返回ndarray形式

t = a.unsqueeze(dim=0)
t = t.permute((1,0)) #互换维度，返回一个新的视图
print(t.shape)
a.float() #等价于.to(torch.float32)
print(a)

# view:返回一个有相同数据但大小不同的tensor。
# 返回的tensor必须有与原tensor相同的数据和相同数目的元素，但可以有不同的大小。
# 一个tensor必须是连续的contiguous()才能被view。
# 和reshape功能类似


x = torch.randn(4, 4)
x.size()
y = x.view(16)

y.size()
z = x.reshape([-1,8])
print(z.size())


#expand:返回tensor的一个新视图，单个维度扩大为更大的尺寸
#tensor也可以扩大为更高维，新增加的维度将附在前面。
#扩大tensor不需要分配新内存，只是仅仅新建一个tensor的视图，
#任何一个大小为1的维度在不分配新内存情况下可扩展为任意的数值。
x = torch.tensor([[1], [2], [3]])
x.size()
t = x.expand(3, 4)
t = t.unsqueeze(dim=0)
t = t.expand(3,-1,-1)

other = torch.zeros([3,4])
x.expand_as(other)

#沿着指定的维度重复tensor。
#不同于expand()，本函数复制的是tensor中的数据
x = torch.tensor([1, 2, 3])
x.repeat(4, 2)  #在x轴复制4次，在y轴复制2次
x.repeat(4, 2, 1).size() #shape->4,2,3

#等价于torch.gather
#实现index与x在dim维度上的聚合
x.gather(dim=1,index=other)



#将函数应用于tensor中的每一个元素
#该函数只能在CPU tensor中使用，并且不应该用在有较高性能要求的代码块。
x = torch.FloatTensor([[-1, 2, 3], [4, 5, 6]])
x.apply_(lambda y: y+2)






```
